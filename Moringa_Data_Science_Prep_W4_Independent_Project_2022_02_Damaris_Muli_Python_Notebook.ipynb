**Deriving new Attributes.**

```python
#Hint:

#To compute usage, we will need to understand that we have to join successive
#(in time) measures/counters for a given station, as the difference will tell whether a car was picked up, returned, or nothing happened
```

```python
# Creating a new column where we find the difference in the number of bluecars at the counter/station

electric_car_sharing_non_duplicates['Bluecar_Difference'] = electric_car_sharing_non_duplicates['Bluecar counter'].diff()
electric_car_sharing_non_duplicates.head()
```

```python
# Creating a new column where we find the difference in the number of Utilib at the counter/station

electric_car_sharing_non_duplicates['Utilib_Difference'] = electric_car_sharing_non_duplicates['Utilib counter'].diff()
electric_car_sharing_non_duplicates.head()
```

```python
# Creating a new column where we find the difference in the number of Utilib 1.4 counter at the counter/station

electric_car_sharing_non_duplicates['utilib1.4_Difference'] = electric_car_sharing_non_duplicates['Utilib 1.4 counter'].diff()
electric_car_sharing_non_duplicates.head()
```

**1.5 Answering Questions**

1.5.1 Bluecar counter

```python
# select city == Paris

electric_car_sharing_paris =electric_car_sharing_non_duplicates[electric_car_sharing_non_duplicates['City'] =='Paris'] 

electric_car_sharing_paris.head(5)
```

```python
#Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

# positive value - a car was returned

# Negative value - a car was picked
electric_car_sharing_paris[electric_car_sharing_paris['Bluecar_Difference'] < 0].groupby('hour')['hour'].count().sort_values(ascending= False).head(1)
```

```python

# What is the most popular hour for returning cars?

electric_car_sharing_non_duplicates[electric_car_sharing_non_duplicates['Bluecar_Difference'] > 0].groupby('hour')['hour'].count().sort_values(ascending= False).head(1)

#21th hour is the most popular hour for returning cars. Its the hour with the highest number of slots
```

```python
#What station is station the most popular?
# let's find the station most popular overall?
electric_car_sharing_non_duplicates[(electric_car_sharing_non_duplicates['Kind'] == 'STATION') & (electric_car_sharing_non_duplicates['Status'] == 'ok')].groupby('Public name')['Public name'].count().sort_values(ascending=False).head(1)
```

```python
# What station is the most popular? At the most popular picking hour?
  
electric_car_sharing_non_duplicates[(electric_car_sharing_non_duplicates['Kind'] == 'STATION') & (electric_car_sharing_non_duplicates['Status'] == 'ok') & (electric_car_sharing_non_duplicates['hour'] == 21)].groupby('Public name')['Public name'].count().sort_values(ascending= False).head(1)


   
```

```python
#  What postal code is the most popular for picking up Bluecars?Overall? Does the most popular station belong to that postal code?


# let's find the most popular postal code for picking up Bluecars

electric_car_sharing_non_duplicates.groupby('Postal code')['Bluecar counter'].count().sort_values(ascending=False).head(1)


#Does the most popular station belong to that postal code?
#Yes
#Uncomment the below code 
#It returns Paris/Porte de Montrouge/8 which is the popular station belongs to postal code  75015
#electric_car_sharing[(electric_car_sharing['Kind'] == 'STATION') & (electric_car_sharing['Status'] == 'ok')].groupby('Postal code')['Postal code'].count().sort_values(ascending=False).head(1)
```

```python
#  What postal code is the most popular for picking up Bluecars at the picking hour? Does the most popular station belong to that postal code?

electric_car_sharing_non_duplicates[(electric_car_sharing_non_duplicates['Kind'] == 'STATION') & (electric_car_sharing_non_duplicates['Status'] == 'ok') &
                                    (electric_car_sharing_non_duplicates['hour'] == 21)].groupby('Postal code')['Bluecar counter'].count().sort_values(ascending=False).head(1)
#No the popular the popular statation does not belong to postal code 75014 
```
